```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sensor_1 = pd.read_csv('01_01mins.csv')
sensor_2 = pd.read_csv('02_01mins.csv')
sensor_3 = pd.read_csv('03_01mins.csv')
sensor_4 = pd.read_csv('04_01mins.csv')
```

Compare PM2_5 Acrross Sensors

Visualizing the first quartile of the datasets

```python
n_rows1 = int(len(sensor_1) * 0.25)
n_rows2 = int(len(sensor_2) * 0.25)
n_rows3 = int(len(sensor_3) * 0.25)
n_rows4 = int(len(sensor_4) * 0.25)
```

```python
plt.figure(figsize=(14, 6))
plt.plot(sensor_1.index[:n_rows1], sensor_1['PM2_5.ug.m3.'][:n_rows1], label='Sensor 1')
plt.xlabel('Time (or Sample Index)')
plt.ylabel('PM2.5 Concentration')
plt.title('PM2.5 in Sensor 1 (1st Quartile)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(sensor_2.index[:n_rows2], sensor_2['PM2_5.ug.m3.'][:n_rows2], color ='orange', label ='Sensor 2')
plt.xlabel('Time (or Sample Index)')
plt.ylabel('PM2.5 Concentration')
plt.title('PM2.5 in Sensor 2 (1st Quartile)')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(14, 6))
plt.plot(sensor_3.index[:n_rows3], sensor_3['PM2_5.ug.m3.'][:n_rows3], color = 'green', label='Sensor 3')
plt.xlabel('Time (1_min)')
plt.ylabel('PM2.5 Concentration')
plt.title('PM2.5 in Sensor 3 (1st Quartile)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(sensor_4.index[:n_rows4], sensor_4['PM2_5.ug.m3.'][:n_rows4], color='red', label='Sensor 4 (25% of data)')
plt.xlabel('Time (or Sample Index)')
plt.ylabel('PM2.5 Concentration')
plt.title('PM2.5 Comparison for Sensor 4 (1st Quartile)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(sensor_1.index[:n_rows1], sensor_1['PM2_5.ug.m3.'][:n_rows1], label ='Sensor 1')
plt.plot(sensor_2.index[:n_rows2], sensor_2['PM2_5.ug.m3.'][:n_rows2], label ='Sensor 2')
plt.plot(sensor_3.index[:n_rows3], sensor_3['PM2_5.ug.m3.'][:n_rows3], label ='Sensor 3')
plt.plot(sensor_4.index[:n_rows4], sensor_4['PM2_5.ug.m3.'][:n_rows4], label ='Sensor 4')
plt.xlabel('Time (or Sample Index)')
plt.ylabel('PM2.5 Concentration')
plt.title('PM2.5 Comparison Across Sensors')
plt.legend()
plt.show()
```
